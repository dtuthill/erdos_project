```python
path = '../Data/UK/cycling_safety_uk_gov.csv'

df = pd.read_csv(path)
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
#Time column has some nans in it, must drop those entries

def is_str(x):
    Nx = len(x)
    I = np.ones(Nx).astype(bool)
    for ii in range(Nx):
        if type(x[ii]) != str:
            I[ii] = False
    return I

df = df.loc[is_str(df.Time.values),:]
```

```python
#make hour, year and month columns
    #hour will be rounded 
def get_hour(T):
    Nt = len(T)
    H = np.zeros(Nt).astype(int)
    for ii in range(Nt):
        s = T[ii].split(':')
        h = int(s[0])
        m = int(s[1])
        if m > 30:
            h += 1
        H[ii] = h % 24
    return H

# def get_ymd(D):
#     Nd = len(D)
#     y = np.zeros(Nd).astype(int)
#     m = np.zeros(Nd).astype(int)
#     d = np.zeros(Nd).astype(int)
#     for ii in range(Nd):
#         s = D[ii].split('-')
#         y[ii] = int(s[0])
#         m[ii] = int(s[1])
#         d[ii] = int(s[2])
#     return y,m,d

def get_ymd(D):
    Nd = len(D)
    # |
    ymd = np.zeros((Nd,3)).astype(int)
    for ii in range(Nd):
        s = D[ii].split('-')
        for nn in range(3):
            ymd[ii,nn] = int(s[nn])
        # ymd[ii,0] = int(s[0])
        # ymd[ii] = int(s[1])
        # d[ii] = int(s[2])
    return ymd
```

```python
df[['y','m','d']] = get_ymd(df.Date.values)
df['h'] = get_hour(df.Time.values)
```

```python
#finally, filter data out for a partciular geographic location. Here we choose Greater London

#M25 box
# Lat_r = [51.348,51.629]
# Lon_r = [-0.458,0.167]

#central London box
Lat_r = [51.475,51.543]
Lon_r = [-0.179,0.023]

I_lat = (df.Latitude > Lat_r[0]) & (df.Latitude < Lat_r[1])
I_lon = (df.Longitude > Lon_r[0]) & (df.Longitude < Lon_r[1])
I = I_lat & I_lon

df_lond = df.loc[I,:].copy()
```

this shows central London



#### Make neighborhoods as a feature 

```python
#24.5/13.25 (this is aspect ratio of above approx. central London selection we made)

#Make a 3 (Lat, N/S) x 6 (Lon, E/W) grid
Nlat = 3
Nlon = 6
Nneigh = Nlon*Nlat

edges_lat = np.linspace(*Lat_r,Nlat+1)
dlat = np.diff(edges_lat)[0]
bins_lat = edges_lat[:-1]+0.5*dlat
edges_lon = np.linspace(*Lon_r,Nlon+1)
dlon = np.diff(edges_lon)[0]
bins_lon = edges_lon[:-1]+0.5*dlon

coord_neigh = [None]*Nneigh

df_lond['neigh'] = 0 #initialize neighborhood feature
ir = 0 #these are lat bins (rows)
ic = 0 #lon bins
for ii in range(Nneigh):
    Ilat = logic_win(df_lond.Latitude,(edges_lat[ir],edges_lat[ir+1]))
    Ilon = logic_win(df_lond.Longitude,(edges_lon[ic],edges_lon[ic+1]))
    I = Ilat & Ilon
    df_lond.loc[I,'neigh'] = ii
    
    coord_neigh[ii] = np.array([bins_lat[ir],bins_lon[ic]])
    
    ic+=1
    if ic == Nlon:
        ic = 0
        ir+=1
```

```python
#make some histograms. Normalize since dataset is pretty unbalanced

feats = ['Speed_limit','h','m','Day_of_Week','neigh','Junction_Control']

nr = 2
nc = 3
fig,ax = plt.subplots(nr,nc,figsize=(5*nc,4*nr))

I1 = df_lond.slight == 1
I0 = df_lond.slight == 0

ir=0
ic=0
for f in feats:
    tup = ax[ir,ic].hist(df_lond.loc[I1,f].values,density=True,rwidth=0.5,color='blue',alpha=0.8,
                  label=1)
    
    ax[ir,ic].hist(df_lond.loc[I0,f].values,density=True,rwidth=0.5,hatch='\\',color='red',alpha=0.4,
                  label=0,bins=tup[1],edgecolor='black')
    
    ax[ir,ic].legend(title='slight')
    ax[ir,ic].set_xlabel(f)
    
    ic+=1
    if ic == nc:
        ic=0
        ir+=1
        
plt.show()
```

```python
#plot neighborhoods again with controlled binning

f = 'neigh'

bins = np.arange(Nneigh)
edges = np.append(bins-0.5,bins[-1]+0.5)

fig,ax = plt.subplots(figsize=(5,4))
ax.hist(df_lond.loc[I1,f].values,density=True,rwidth=0.5,color='blue',alpha=0.8,
                  label=1,bins=edges)
    
ax.hist(df_lond.loc[I0,f].values,density=True,rwidth=0.5,hatch='\\',color='red',alpha=0.4,
              label=0,edgecolor='black',bins=edges)

ax.legend(title='slight')
ax.set_xlabel(f)
ax.set_xticks(np.arange(Nneigh).astype(int))

plt.show()
```
